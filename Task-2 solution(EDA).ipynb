# Line Chart (Trend)

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
monthly_sales = df.groupby(
    df['Date'].dt.to_period('M')
)['TotalPrice'].sum()
monthly_sales.plot(
    kind = 'line',
    figsize = (10,5),
    marker ='o'
)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
monthly_order = df.groupby(df['Date'].dt.to_period('M'))['Quantity'].sum()
monthly_order
monthly_order.plot(
    kind = 'line',
    figsize = (10,5),
    marker ='o'
)
plt.title('Monthly Order Trend')
plt.xlabel('Month')
plt.ylabel('Total orders')
plt.show()
```

# Revenue By Product

```python
product_sales = df.groupby(
    'Product'
)['TotalPrice'].sum().sort_values(
    ascending = False
)

product_sales.plot(
    kind = 'bar',
    figsize = (10,5)
)
plt.show()
```

# Coupon Analysis

```python
# coupon usage counts
df['CouponCode'].value_counts()


#Observation:
SAVE10 is the most frequently used coupon among customers.
```

```python
# revenue by coupon 
df.groupby('CouponCode')['TotalPrice'].sum()


#Observation:
#Orders associated with SAVE20 generated the highest total revenue.
```

```python
#average quantity by coupon 
df.groupby('CouponCode')['Quantity'].mean()


#Observation:
#Customers using discount coupons tend to purchase slightly more items per order.
```

```python
# order counts by coupon 
df.groupby('CouponCode')['OrderID'].count()


#Observation:
#A significant proportion of customers completed purchases without using a coupon.
```
